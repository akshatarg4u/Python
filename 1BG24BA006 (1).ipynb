```python
dt2 = {"Book_ID": [101, 102, 103, 104],
       "Book_Name": ["Python Basics", "Data Structures", "DBMS", "Operating Systems"],
       "Authors": ["Guido van Rossum", "Rance D. Necaise", "Raghu Ramakrishnan", "Abraham Silberschatz"],
       "Status":["Available", "Issued", "Available", "Available"]}
dt2
```

```python
def show_menu():
    dt2 = {
        "Book_ID": [],
        "Book_Name": [],
        "Author_Name": [],
        "Status": []
    }

    while True:
        print("\n--- Library Management Menu ---")
        print("1. View Book")
        print("2. Add Book")
        print("3. Search Book by ID")
        print("4. Delete Book")
        print("5. Exit")

        choice = input("Enter your choice: ")

        # View Library
        if choice == "1":
            if len(dt2["Book_ID"]) == 0:
                print("No books available.")
            else:
                print("\nLibrary Details:")
                for i in range(len(dt2["Book_ID"])):
                    print(
                        dt2["Book_ID"][i],
                        dt2["Book_Name"][i],
                        dt2["Author_Name"][i],
                        dt2["Status"][i]
                    )

        # Add Book
        elif choice == "2":
            try:
                dt2["Book_ID"].append(int(input("Enter Book ID: ")))
                dt2["Book_Name"].append(input("Enter Book Name: ").strip())
                dt2["Author_Name"].append(input("Enter Author Name: ").strip())
                dt2["Status"].append("Available")   # default status

                print("Book added successfully!")
            except ValueError:
                print("Invalid input! Book ID must be a number.")

        # Search Book
        elif choice == "3":
            try:
                search_id = int(input("Enter Book ID to search: "))
                if search_id in dt2["Book_ID"]:
                    index = dt2["Book_ID"].index(search_id)
                    print("Book Found:")
                    print(
                        dt2["Book_ID"][index],
                        dt2["Book_Name"][index],
                        dt2["Author_Name"][index],
                        dt2["Status"][index]
                    )
                else:
                    print("Book not found.")
            except ValueError:
                print("Invalid Book ID!")

        # Delete Book
        elif choice == "4":
            try:
                del_id = int(input("Enter Book ID to delete: "))
                if del_id in dt2["Book_ID"]:
                    index = dt2["Book_ID"].index(del_id)
                    for key in dt2:
                        dt2[key].pop(index)
                    print("Book deleted successfully!")
                else:
                    print("Book not found.")
            except ValueError:
                print("Invalid Book ID!")

        # Exit
        elif choice == "5":
            print("Exiting Library System...")
            break

        else:
            print("Invalid choice! Please try again.")


# Function call
show_menu()
```
