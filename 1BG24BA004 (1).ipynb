```python
# Expense Tracker
expense = {
    "Expense_ID": [],
    "Category": [],
    "Amount": [],
    "Note": []}

while True:
    print("\n--- Expense Tracker ---")
    print("1. View Expense")
    print("2. Add Expense")
    print("3. Exit")

    choice = input("Enter choice: ")

    # View Expenses
    if choice == "1":
        for i in range(len(expense["Expense_ID"])):
            print(
                expense["Expense_ID"][i],
                expense["Category"][i],
                expense["Amount"][i],
                expense["Note"][i])

    # Add Expense
    elif choice == "2":
        expense["Expense_ID"].append(len(expense["Expense_ID"]) + 1)
        expense["Category"].append(input("Enter Category: "))
        expense["Amount"].append(float(input("Enter Amount: ")))
        expense["Note"].append(input("Enter Note: "))

        print("Expense added!")

    # Exit
    elif choice == "3":
        print("Exiting Expense Tracker...")
        break

    else:
        print("Invalid choice!")
```
